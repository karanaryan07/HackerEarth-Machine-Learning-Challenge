```python
import pandas as pd
import numpy as np
```

```python
dataset_train = pd.read_csv('train.csv')
```

```python
dataset_train.head()
```

```python
x = dataset_train.drop(['Severity'] , axis = 1)
```

```python
y = dataset_train['Severity']
```

```python
x = x.drop(['Accident_ID'], axis = 1)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size = 0.15 , random_state = 0 )
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test  = sc_x.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10 , criterion = 'entropy')
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
cm = accuracy_score(y_test , y_pred)
```

```python
import tensorflow as tf
```

```python
from keras import Sequential
```

```python
from keras.layers import Dense
```

```python
classier = Sequential()
```

```python
classifier = tf.keras.Sequential([
    tf.keras.layers.Dense(500 , activation = 'relu' , input_dim = 10),
    tf.keras.layers.Dense(300 , activation = 'relu'),
    tf.keras.layers.Dense(100 , activation = 'relu'),
    tf.keras.layers.Dense(50 , activation = 'relu'),
    tf.keras.layers.Dense(4, activation = 'softmax')
])
```

```python
classifier.summary()
```

```python
classifier.compile(loss = 'sparse_categorical_crossentropy',
                 optimizer = 'adam',
                 metrics = ['accuracy'])
```

```python
classifier.fit(x_train , y_train, batch_size = 16, epochs = 10 , validation_data = (x_test , y_test))
```
